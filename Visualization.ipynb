# Распределение зарплаты на HeadHunter

```python
def SalaryVisual(columns):
    f,axis = plt.subplots(8,3, figsize = (10,20))
    plt.tight_layout(pad=1.0, w_pad=4.0, h_pad=4.0)
    i,j = 0,0
    for language in range(0,len(columns),200):
        exp_lang = columns[language:language+200]
        axis[i,j].set_xlabel('Serial number')
        axis[i,j].set_ylabel('Salary')
        axis[i,j].plot(exp_lang)
        axis[i,j].set_title('Salary distribution')
        if j == 2: 
            j=0
            i+=1
        else: j +=1
        if i== 8: break
```

```python
SalaryVisual(hdt["Salary"].values)
```
